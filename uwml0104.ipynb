```
# Word counter function
def word_count(str):
  counts = {}
  words = str.split()
  for word in words:
    if word in counts:
      counts[word] += 1
    else:
      counts[word] = 1
    #counts = collections.OrderedDict(counts)
  return counts
```

```
people_wiki = people_wiki_raw
people_wiki['word_count'] = people_wiki['text'].apply(word_count)
people_wiki
```

## Test TF-IDF vectorizer example

```
data = [
['DOC1', 'is month lenders loan information on loan'],
['DOC2', 'loan information is important is loan loan'],
['DOC3', 'loan must always submit loan information month'],
['DOC4', 'tesla is considered is overvalued stock present'],
['DOC5', 'who thinks tesla is overvalued is present'],
]

df = pd.DataFrame(data, columns = ['doc', 'text'])

df
```

```
tvec = feature_extraction.text.TfidfVectorizer(smooth_idf=False)
tfidf_mtrx = tvec.fit_transform(df['text'])
```

```
tvec.vocabulary_
```

```
# Full TF-IDF sparse matrix
print(tfidf_mtrx)
```

```
# Document 1
print(tfidf_mtrx[0])
```

```
print(tfidf_mtrx.toarray())
```

```
tvec.get_feature_names()
```

```
# Find document scores
indx = df[df['doc'] == 'DOC1'].index.tolist()
tfidf_array = tfidf_mtrx.toarray()[indx][0]

dict(zip(tvec.get_feature_names(), tfidf_array))
```

```
#test = tfidf_mtrx.toarray()[indx]
test = tfidf_mtrx.toarray()[0]
#print(type(test))
test
```

```
# Data frame solution
rslt_df = pd.DataFrame(tfidf_mtrx.toarray(), columns = tvec.get_feature_names())
rslt_df = rslt_df.T
rslt_df.columns = df['doc'].tolist()
rslt_df
```
