```python
# Regression slope and intercept

def simple_linear_regression(input_feature, output):
  y = output.values
  x = input_feature.values
  num = np.mean(x * y) - (np.mean(x) * np.mean(y))
  den = np.mean(x * x) - (np.mean(x) * np.mean(x))
  slope = num/den
  intercept = np.mean(y) - slope * np.mean(x)
  return(intercept, slope)
```

```python
# Test
x = [0,1,2,3,4]
y = [1,3,7,13,21]
x = pd.Series(x)
y = pd.Series(y)

model = simple_linear_regression(x, y)
print(model[0])
print(model[1])
```

```python
train_y = train['price']
train_x = train['sqft_living']
train_x
```

```python
model = simple_linear_regression(train_x, train_y)
squarefeet_intercept = model[0]
squarefeet_slope = model[1]

print(squarefeet_intercept)
print(squarefeet_slope)
```

```python
# Predictions
def get_regression_predictions(input_feature, intercept, slope):
  x = input_feature
  predicted_output = intercept + (slope * x)
  return(predicted_output)
```

```python
# Q1
get_regression_predictions(2650, squarefeet_intercept, squarefeet_slope)
```

```python
# RSS function
def get_residual_sum_of_squares(input_feature, output, intercept, slope):
  x = input_feature.values
  y = output.values
  yhat =  intercept + (slope * x)
  e = yhat - y
  RSS = np.sum(e * e)
  return(RSS)
```

```python
get_residual_sum_of_squares(train['sqft_living'], train['price'], squarefeet_intercept, squarefeet_slope)
```
