```python
test = pd.read_csv('gdrive/My Drive/uwml/kc_house_test_data.csv')
train = pd.read_csv('gdrive/My Drive/uwml/kc_house_train_data.csv')
```

```python
# New attributes
train['bedrooms_squared'] = train['bedrooms']**2
train['bed_bath_rooms'] = train['bedrooms'] * train['bathrooms']
train['log_sqft_living'] = np.log(train['sqft_living'])
train['lat_plus_long'] = train['lat']+train['long']

test['bedrooms_squared'] = test['bedrooms']**2
test['bed_bath_rooms'] = test['bedrooms'] * test['bathrooms']
test['log_sqft_living'] = np.log(test['sqft_living'])
test['lat_plus_long'] = test['lat']+test['long']
```

```python
# Test
test.describe()
```

```python
train_y = train['price']
train_x = train[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']]

train_x = sm.add_constant(train_x)
model1 = sm.OLS(train_y, train_x).fit()

print(model1.summary())
```

```python
train_y = train['price']
train_x = train[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms']]

train_x = sm.add_constant(train_x)
model2 = sm.OLS(train_y, train_x).fit()

print(model2.summary())
```

```python
train_y = train['price']
train_x = train[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms', 'log_sqft_living', 'lat_plus_long']]

train_x = sm.add_constant(train_x)
model3 = sm.OLS(train_y, train_x).fit()

print(model3.summary())
```

```python
print(model1.ssr/1e9)
print(model2.ssr/1e9)
print(model3.ssr/1e9)
```

```python
test_y = test['price']
test_x1 = test[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']] 
test_x2 = test[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms']]
test_x3 = test[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms', 'log_sqft_living', 'lat_plus_long']]
```

```python
model1.predict(sm.add_constant(test_x1))
```

```python
# RSS function
def get_residual_sum_of_squares(yhat, y):
  e = yhat - y
  RSS = np.sum(e * e)
  return(RSS)
```

```python
print(get_residual_sum_of_squares(model1.predict(sm.add_constant(test_x1)), test_y)/1e9)
print(get_residual_sum_of_squares(model2.predict(sm.add_constant(test_x2)), test_y)/1e9)
print(get_residual_sum_of_squares(model3.predict(sm.add_constant(test_x3)), test_y)/1e9)
```
